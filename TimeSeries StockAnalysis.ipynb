# Time Series Data là gì và các loại của nó?

**Trước khi chuyển ngay sang Time Series Analysis, trước tiên chúng ta hãy hiểu Time Series Analysis là gì.**

* Time Series là tập hợp các điểm dữ liệu trong một khoảng thời gian nhất định. Vẽ các điểm trên biểu đồ và một trong các trục của bạn sẽ luôn là thời gian.
* Điều khiến dữ liệu Time Series khác biệt với các dữ liệu khác là việc phân tích có thể cho thấy các biến thay đổi theo thời gian như thế nào.
* Tần suất ghi điểm dữ liệu có thể là hàng giờ, hàng ngày, hàng tuần, hàng tháng, hàng quý hoặc hàng năm.
* Nói cách khác, thời gian là một biến số quan trọng vì nó cho thấy dữ liệu điều chỉnh như thế nào theo tiến trình của các điểm dữ liệu cũng như kết quả cuối cùng.
  Nó cung cấp một nguồn thông tin bổ sung và một thứ tự phụ thuộc được thiết lập giữa các dữ liệu.

* Dữ liệu chuỗi thời gian có thể có ba loại: -
1. **Time series data** - Việc quan sát các giá trị của một biến được ghi lại tại các thời điểm khác nhau được gọi là dữ liệu chuỗi thời gian.
1. **Cross sectional data** - Là dữ liệu của một hoặc nhiều biến được ghi lại tại cùng một thời điểm.
1. **Pooled data**- Là sự kết hợp giữa dữ liệu chuỗi thời gian và dữ liệu cắt ngang.

# Time Series Analysis là gì?

**Bây giờ chúng ta đã hiểu dữ liệu Time Series nghĩa là gì .. hãy hiểu phân tích Time Series là gì?**
* Phân tích Time Series là phương pháp phân tích dữ liệu để trích xuất các thông tin và đặc điểm thống kê hữu ích.
* Phân tích Time Series bao gồm các phương pháp thống kê để phân tích dữ liệu Time Series. Những phương pháp này cho phép chúng tôi trích xuất các số liệu thống kê, mẫu có ý nghĩa và các đặc điểm khác của dữ liệu.
* Chuỗi thời gian được hiển thị bằng biểu đồ đường. Vì vậy, phân tích Time Series liên quan đến việc hiểu các khía cạnh vốn có của dữ liệu Time Series để chúng ta có thể tạo ra các dự báo chính xác và có ý nghĩa.
* Một trong những mục tiêu chính của nghiên cứu là dự đoán giá trị trong tương lai.

Ví dụ về phân tích Time Series:

# Tại sao các tổ chức sử dụng phân tích dữ liệu time series ?

* Phân tích Time series giúp tổ chức hiểu được nguyên nhân cơ bản của các xu hướng hoặc mô hình hệ thống theo thời gian.
* Bằng cách sử dụng trực quan hóa dữ liệu, người dùng doanh nghiệp có thể thấy xu hướng theo mùa và tìm hiểu sâu hơn lý do tại sao những xu hướng này lại xảy ra.
* Khi các tổ chức phân tích dữ liệu theo các khoảng thời gian nhất quán, họ cũng có thể sử dụng tính năng dự báo Time series để dự đoán khả năng xảy ra các sự kiện trong tương lai.

# Các loại phân tích chuỗi thời gian

Một số mô hình phân tích time series bao gồm -

**1 Classification**: Nó xác định và gán các danh mục cho dữ liệu.

**2 Curvefitting**: Nó vẽ dữ liệu trên một đường cong để nghiên cứu mối quan hệ giữa các biến trong dữ liệu.

**3 Descriptive Analysis**: Các mẫu trong dữ liệu time-series, chẳng hạn như xu hướng, chu kỳ và biến đổi theo mùa, được xác định.

**4 Explanative analysis**: Nó cố gắng hiểu dữ liệu và mối quan hệ giữa nó với nguyên nhân và kết quả.

**5 Segmentation**: Nó chia dữ liệu thành các phân đoạn để hiển thị các thuộc tính cơ bản của dữ liệu nguồn.

# Các thành phần của chuỗi thời gian

1. **Trend** - Xu hướng hiển thị hướng chung của dữ liệu chuỗi thời gian trong một khoảng thời gian dài. Xu hướng có thể tăng (tăng), giảm (giảm) hoặc ngang (đứng yên).
1. **Seasonality** - Thành phần tính thời vụ thể hiện xu hướng lặp lại về thời gian, hướng và cường độ. Một số ví dụ bao gồm sự gia tăng lượng nước tiêu thụ vào mùa hè do điều kiện thời tiết nóng bức.
1. **Noise** - Giá trị ngoại lệ hoặc giá trị bị thiếu
1. **Cyclical Component** - Đây là những xu hướng không lặp lại cố định trong một khoảng thời gian cụ thể. Chu kỳ đề cập đến khoảng thời gian thăng trầm, bùng nổ và suy thoái của một chuỗi thời gian, chủ yếu được quan sát thấy trong các chu kỳ kinh doanh. Các chu kỳ này không thể hiện sự thay đổi theo mùa nhưng thường xảy ra trong khoảng thời gian từ 3 đến 12 năm tùy thuộc vào tính chất của chuỗi thời gian.
1. **Irregular Variation** - Đây là những biến động trong dữ liệu chuỗi thời gian trở nên rõ ràng khi loại bỏ các biến động theo xu hướng và chu kỳ. Những biến thể này không thể đoán trước, thất thường và có thể ngẫu nhiên hoặc không.
1. **ETS Decomposition** - Phân tách ETS được sử dụng để phân tách các thành phần khác nhau của chuỗi thời gian. Thuật ngữ ETS là viết tắt của Lỗi, Xu hướng và Tính thời vụ.

# Stationary Test / ADF (Augmented Dickey-Fuller) Test

Chuỗi thời gian dừng là chuỗi có các đặc tính thống kê như giá trị trung bình, phương sai, tự tương quan, v.v. đều không đổi theo thời gian.

*  Strong stationarity: là quá trình ngẫu nhiên mà phân bố xác suất chung vô điều kiện không thay đổi khi dịch chuyển theo thời gian. Do đó, các tham số như giá trị trung bình và phương sai cũng không thay đổi theo thời gian.
*  Weak stationarity: là quá trình trong đó giá trị trung bình, phương sai, tự tương quan không đổi trong suốt thời gian

Stationarity là quan trọng vì chuỗi không cố định phụ thuộc vào thời gian có quá nhiều tham số cần tính đến khi lập mô hình chuỗi thời gian. Phương thức diff() có thể dễ dàng chuyển đổi một chuỗi không cố định thành một chuỗi đứng yên.

* Đầu tiên, chúng ta cần kiểm tra xem một chuỗi có tính dừng hay không vì phân tích chuỗi thời gian chỉ thực hiện được với dữ liệu dừng.
* Bài kiểm tra Dickey-Fuller là một trong những bài kiểm tra thống kê phổ biến nhất.
* Nó có thể được sử dụng để xác định sự hiện diện của nghiệm đơn vị trong chuỗi và do đó giúp chúng ta hiểu liệu chuỗi có dừng hay không. Giả thuyết vô giá trị và thay thế của thử nghiệm này là:

**Null Hypothesis**: Chuỗi có nghiệm đơn vị (giá trị a =1)

**Alternate Hypothesis***: Chuỗi không có nghiệm đơn vị.

Nếu chúng ta không bác bỏ giả thuyết không, chúng ta có thể nói rằng chuỗi này không dừng. Điều này có nghĩa là chuỗi có thể là tuyến tính hoặc sai phân.

**Nếu cả giá trị trung bình và độ lệch chuẩn đều là các đường thẳng (trung bình không đổi và phương sai không đổi), thì chuỗi sẽ dừng.**

```python
##Kiểm tra tính dừng
def test_stationarity(timeseries):

    #Xác định số liệu thống kê lăn
    rolmean = timeseries.rolling(12).mean()
    rolstd = timeseries.rolling(12).std()

    #Plot thống kê cuộn:
    plt.figure(figsize=(15,5))
    plt.plot(timeseries,color='blue',label='Original')
    plt.plot(rolmean,color='red',label='Rolling Mean')
    plt.plot(rolstd, color='black', label = 'Rolling Std')
    plt.legend(loc='best')
    plt.title('Giá trị trung bình lăn và độ lệch chuẩn')
    plt.show(block=False)

    print("Results of dickey fuller test")
    adft = adfuller(timeseries,autolag='AIC')
# đầu ra cho dft sẽ cung cấp cho chúng ta mà không cần xác định giá trị là gì.
     #do đó chúng tôi viết thủ công những giá trị nào nó giải thích bằng vòng lặp for
    output = pd.Series(adft[0:4],index=['Test Statistics','p-value','No. of lags used','Number of observations used'])
    print(output)

test_stationarity(data['Close'])
```

Thông qua biểu đồ trên, chúng ta có thể thấy giá trị trung bình và độ lệch chuẩn ngày càng tăng và do đó **chuỗi của chúng ta không dừng.**

Chúng tôi thấy rằng giá trị p lớn hơn 0,05 nên chúng tôi không thể bác bỏ giả thuyết Null. Ngoài ra, số liệu thống kê kiểm tra lớn hơn các giá trị quan trọng. nên dữ liệu không cố định.

# ARIMA Model

* Autoregressive integrated moving average (ARIMA) dự đoán các giá trị trong tương lai dựa trên các giá trị trong quá khứ.
* ARIMA sử dụng các đường trung bình động có độ trễ để làm mượt dữ liệu chuỗi thời gian.
* Chúng được sử dụng rộng rãi trong phân tích kỹ thuật để dự báo giá chứng khoán trong tương lai.

  Theo tên gọi, chúng ta có thể chia mô hình thành các thành phần nhỏ hơn như sau:
1. **AR**: mô hình AutoregRegressive đại diện cho một loại quy trình ngẫu nhiên. Đầu ra của mô hình phụ thuộc tuyến tính vào giá trị trước đó của chính nó, tức là một số điểm dữ liệu bị trễ hoặc số lượng quan sát trong quá khứ.
1. **I**: được tích hợp ở đây có nghĩa là bước sai phân để tạo ra dữ liệu chuỗi thời gian cố định, tức là loại bỏ các thành phần theo mùa và xu hướng.
1. **MA**: mô hình Trung bình trượt có đầu ra phụ thuộc tuyến tính vào các quan sát hiện tại và các quan sát khác nhau trong quá khứ của một số hạng ngẫu nhiên.

Mô hình ARIMA thường được ký hiệu là **ARIMA(p, d, q)** và tham số p, d, q được định nghĩa như sau:

1. **p**: thứ tự độ trễ hoặc số độ trễ thời gian của mô hình tự hồi quy AR(p)
1. **d**: mức độ khác biệt hoặc số lần dữ liệu bị trừ đi giá trị quá khứ
1. **q**: thứ tự của mô hình trung bình động MA(q)

 Giá trị của p, d và q được chọn bằng cách quan sát các biểu đồ của ACF và PACF nhưng ngay bây giờ chúng ta sẽ sử dụng Auto ARIMA để có được các tham số tốt nhất mà không cần vẽ biểu đồ ACF và PACF.


**Auto ARIMA**: Tự động khám phá thứ tự tối ưu cho mô hình ARIMA.

Hàm auto_arima tìm cách xác định các tham số tối ưu nhất cho mô hình ARIMA và trả về mô hình ARIMA phù hợp. Hàm này dựa trên hàm R thường được sử dụng, dự báo::auto.arima.

 ## ARIMA

- Mô hình ARIMA (Autoregressive Integrated Moving Average) là một mô hình dùng để mô tả và dự đoán chuỗi thời gian. Nó kết hợp ba thành phần chính: AR (Autoregressive), I (Integrated), và MA (Moving Average), để mô tả sự biến đổi và dự đoán trong dữ liệu chuỗi thời gian.

Dưới đây là bản chất của mỗi thành phần trong mô hình ARIMA:

1. AR (Autoregressive): Đây là phần tự hồi quy của mô hình. Nó đo lường tương quan giữa các giá trị trước đó trong chuỗi thời gian. ARIMA sử dụng p-lags (được gọi là "p-order") để xác định số lượng lags trước đó mà mô hình cần xem xét. Thành phần AR giúp mô hình bắt chước tương quan trong dữ liệu.

2. I (Integrated): Đây là phần thực hiện chuyển đổi sai phân của chuỗi thời gian. Dự đoán chuỗi thời gian thường dựa trên giả định rằng nó là dễ dàng dự đoán nếu chuỗi thời gian là dễ dự đoán. Do đó, thành phần này thực hiện sai phân d (được gọi là "d-order") để biến đổi chuỗi thành một dạng dễ dự đoán hơn.

3. MA (Moving Average): Phần này dự đoán giá trị dựa trên một số lượng quá trình trung bình động của sai phân chuỗi thời gian. MA sử dụng q-lags (được gọi là "q-order") để xác định số lượng quá trình trung bình động cần xem xét.

- Bản chất của mô hình ARIMA là kết hợp cả ba thành phần này để mô phỏng và dự đoán chuỗi thời gian. Dự đoán ARIMA xảy ra bằng cách sử dụng các thông tin về lịch sử (AR), sai phân (I), và các thông tin về quá trình trung bình động (MA). Mô hình ARIMA cố gắng khớp chuỗi dữ liệu thực tế bằng cách tối ưu hóa các thông số p, d, và q để có kết quả dự đoán tốt nhất.

- Mô hình ARIMA là một công cụ mạnh mẽ trong dự đoán chuỗi thời gian và phân tích dữ liệu thời gian, và nó đã được ứng dụng rộng rãi trong nhiều lĩnh vực khác nhau.

```python
import warnings
import math
import pandas as pd
import plotly.graph_objects as go
from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import mean_squared_error, mean_absolute_error
from tqdm import tqdm

# Assuming 'data' is a pandas DataFrame with a 'Close' column and DateTime index.

# Filtering data for the required timeframe
stocks_data = data[data.index > "2021"]
df_train = stocks_data[stocks_data.index < "2023"]
df_valid = stocks_data[stocks_data.index >= "2023"]
train_data = df_train['Close'].values
test_data = df_valid['Close'].values

def train_test_split(data):
    train_size = int(len(data) * 0.7)
    return data[:train_size], data[train_size:]

def arima_model(history, y, order=(5, 0, 5)):
    predictions = []
    model_fit = None
    for i in tqdm(range(len(y))):
        model = ARIMA(history, order=order)
        model_fit = model.fit()
        result = model_fit.forecast()[0]
        predictions.append(result)
        history.append(y[i])
    return predictions, model_fit.summary() if model_fit else None

# Splitting the data
train_data, test_data = train_test_split(stocks_data['Close'])

# ARIMA Model Prediction
history = [x for x in train_data]
predictions, report_model = arima_model(history, test_data)

# Evaluation Metrics
mse = mean_squared_error(test_data, predictions)
mae = mean_absolute_error(test_data, predictions)
rmse = math.sqrt(mse)

# Display Metrics
print('MSE:', mse)
print('MAE:', mae)
print('RMSE:', rmse)

# Optionally, print the ARIMA model summary
print(report_model)

# Plotting with Plotly
fig = go.Figure()

# Train Stock Price
fig.add_trace(go.Scatter(x=df_train.index, y=df_train['Close'], mode='lines', name='Train Stock Price', line=dict(color='green')))

# Real Stock Price
fig.add_trace(go.Scatter(x=df_valid.index, y=df_valid['Close'], mode='lines', name='Real Stock Price', line=dict(color='red')))

# Predicted Stock Price
min_length = min(len(predictions), len(df_valid.index))
fig.add_trace(go.Scatter(x=df_valid.index[:min_length], y=predictions[:min_length], mode='lines', name='Predicted Stock Price', line=dict(color='blue')))

fig.update_layout(title='Stock Price Prediction by ARIMA', xaxis_title='Time', yaxis_title='Stock Price')

fig.show()
```
